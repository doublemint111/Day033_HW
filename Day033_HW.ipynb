```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://www.ptt.cc/bbs/Gossiping/index.html', verify = False)
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.text)
```

模仿授權機制
方法一：先送登入，再送請求

```python
rs = requests.session()
payload={
    'from':'/bbs/Gossiping/index.html',
    'yes': 'yes'
}
res = rs.post('https://www.ptt.cc/ask/over18',verify = False, data = payload)
res = rs.get('https://www.ptt.cc/bbs/Gossiping/index.html',verify = False)
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text) 
```

方法二：在請求中帶入 cookie

```python
res = requests.get('https://www.ptt.cc/bbs/Gossiping/index.html',verify = False, cookies={'over18': '1'})
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text)
```

```python
res = requests.get('https://github.com/new',verify = False, cookies={'over18': '1'})
soup = BeautifulSoup(res.text,'html.parser')
soup.title
```

```python
cookies = {
    'logged_in': 'yes',
    'user_session': '...'
}

r = requests.get('https://github.com/new', cookies=cookies)
soup = BeautifulSoup(r.text, 'lxml')
soup.title # 加上檢查的 cookies ，就可以正確取得資料
```
